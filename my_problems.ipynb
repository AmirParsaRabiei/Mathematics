```python
from sympy import Matrix, solve_linear_system, symbols

# Define the symbolic variables
x, y = symbols('x y')

# Create the coefficient matrices
A = Matrix([[1, 1], [0, 1]])
b = Matrix([0, 0])

# Solve the linear system
solution = solve_linear_system(A, b)

# Print the solution
print("Solution:")
print(f"x = {solution[x]}")
print(f"y = {solution[y]}")
```

```python
import numpy as np

# Define the coefficients matrix A
A = np.array([[0, 1], [0, 3]], dtype=np.dtype(float))

# Define the constants vector b
b = np.array([[0],[2]], dtype=np.dtype(float))

# Determinate coefficients
d = np.linalg.det(A)
print(f"Determinant of matrix A: {d:.2f}\n")

try:
    # Solve the system of linear equations
    solution = np.linalg.solve(A, b)

    # Print the solution
    print("Solution:")
    print("x =", solution[0])
    print("y =", solution[1])
except np.linalg.LinAlgError as err:
    print(err)
```

```python
import numpy as np

# Define the coefficients matrix A
A = np.array([[2, 1], [0, 3]], dtype=np.dtype(float))

# Define the constants vector b
b = np.array([[2], [2]], dtype=np.dtype(float))

# Create the augmented matrix [A|b]
augmented_matrix = np.concatenate((A, b), axis=1)

# Calculate the rank of the coefficient matrix A
rank_A = np.linalg.matrix_rank(A)

# Calculate the rank of the augmented matrix [A|b]
rank_augmented = np.linalg.matrix_rank(augmented_matrix)

print(f"Rank of coefficient matrix A: {rank_A}")
print(f"Rank of augmented matrix [A|b]: {rank_augmented}\n")

if rank_A == rank_augmented:
    if rank_A == min(A.shape):
        print("The system has a unique solution.")
        try:
            # Solve the system of linear equations
            solution = np.linalg.solve(A, b)

            # Print the solution
            print("Solution:")
            print("x =", solution[0, 0])
            print("y =", solution[1, 0])
        except np.linalg.LinAlgError as err:
            print(err)
    else:
        print("The system has infinitely many solutions.")
        
        # Find the pivot variables and free variables
        pivots = []
        free_vars = []
        for j in range(A.shape[1]):
            if np.linalg.matrix_rank(A[:, :j+1]) > len(pivots):
                pivots.append(j)
            else:
                free_vars.append(j)
        
        # Express the solution in parametric form
        solution = np.zeros((A.shape[1], 1))
        solution[free_vars, 0] = 1
        solution[pivots, 0] = np.linalg.solve(A[:, pivots], b - A[:, free_vars] @ solution[free_vars, 0].reshape(-1, 1))
        
        print("The solution is of the form:")
        for i in range(A.shape[1]):
            if i in pivots:
                print(f"x{i+1} = {solution[i, 0]:.2f}", end="")
            else:
                print(f"x{i+1} = t", end="")
            
            for j in free_vars:
                if i == j:
                    continue
                if A[i, j] < 0:
                    print(f" + {-A[i, j]:.2f}t", end="")
                elif A[i, j] > 0:
                    print(f" - {A[i, j]:.2f}t", end="")
            print()
else:
    print("The system has no solution.")
```
