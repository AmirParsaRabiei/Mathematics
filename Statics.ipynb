```python
# Differences between Standard Deviation and Variance
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = np.array([2, 4, 4, 4, 5, 5, 7, 9])

# Mean
mean = np.mean(data)

# Variance
variance = np.var(data)

# Standard Deviation
std_dev = np.sqrt(variance)

# Plotting
plt.figure(facecolor='white')
plt.plot(data, 'o', label='Data Points')
plt.axhline(y=mean, color='r', linestyle='-', label='Mean')
plt.axhline(y=mean + std_dev, color='g', linestyle='--', label='Mean + 1 Std Dev')
plt.axhline(y=mean - std_dev, color='g', linestyle='--', label='Mean - 1 Std Dev')
plt.axhline(y=mean + 2*std_dev, color='b', linestyle='--', label='Mean + 2 Std Dev')
plt.axhline(y=mean - 2*std_dev, color='b', linestyle='--', label='Mean - 2 Std Dev')

plt.title('Data Points with Mean and Standard Deviations')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
print(f"Mean={mean} // Standard Deviation={std_dev} // Variance={variance}")
print("Plot generated with mean, standard deviation, and variance lines.")
```

```python
# Gaussian distribution of above plot

import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = np.array([2, 4, 4, 4, 5, 5, 7, 9])

# Mean and Standard Deviation
mean = np.mean(data)
std_dev = np.std(data)

# Generate Gaussian distribution data
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
gaussian = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Plotting
plt.figure(facecolor='white')
plt.plot(x, gaussian, label='Gaussian Distribution')
plt.axvline(x=mean, color='r', linestyle='-', label='Mean')
plt.axvline(x=mean + std_dev, color='g', linestyle='--', label='Mean + 1 Std Dev')
plt.axvline(x=mean - std_dev, color='g', linestyle='--', label='Mean - 1 Std Dev')
plt.axvline(x=mean + 2*std_dev, color='b', linestyle='--', label='Mean + 2 Std Dev')
plt.axvline(x=mean - 2*std_dev, color='b', linestyle='--', label='Mean - 2 Std Dev')

plt.title('Gaussian Distribution with Mean and Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print("Gaussian distribution plot generated with mean and standard deviation lines.")
```

- **Variance** measures the average squared deviation from the mean. Because it squares the deviations, it gives more weight to larger deviations (outliers). This makes variance particularly sensitive to outliers and can be useful when you want to emphasize the impact of extreme values.

- **Standard Deviation** is the square root of the variance. It measures the average deviation from the mean in the same units as the data. Standard deviation is more intuitive and easier to interpret because it is in the same units as the original data. It is commonly used to understand the spread of data and is often preferred for general purposes.

In summary:
- **Variance** is suitable for emphasizing the impact of outliers.
- **Standard Deviation** is generally used for understanding the overall spread of data.

An Applications of Standard Deviation and Variance: 

Finance:

- Risk Assessment: Standard deviation is used to measure the volatility of stock prices or investment returns. A higher standard deviation indicates higher risk.
- Portfolio Management: Variance and standard deviation help in constructing diversified portfolios to minimize risk.



### Decision-Making Implications
- **Risk Assessment**: In finance, a higher variance indicates higher risk due to the potential for extreme values. Standard deviation is often used to assess the volatility of an investment.
- **Quality Control**: In manufacturing, standard deviation is used to monitor process consistency. A lower standard deviation indicates a more consistent process.
- **Research and Experiments**: In scientific research, standard deviation is used to report the variability of data. It helps in understanding the reliability and precision of measurements.

### Example
Let's consider an example with two datasets:
- Dataset A: [10, 12, 14, 16, 18]
- Dataset B: [10, 12, 14, 16, 100]

```python
import numpy as np

# Datasets
dataset_a = np.array([10, 12, 14, 16, 18])
dataset_b = np.array([10, 12, 14, 16, 100])

# Mean
mean_a = np.mean(dataset_a)
mean_b = np.mean(dataset_b)

# Variance
variance_a = np.var(dataset_a)
variance_b = np.var(dataset_b)

# Standard Deviation
std_dev_a = np.sqrt(variance_a)
std_dev_b = np.sqrt(variance_b)

# Print results
print("Dataset A:")
print("Mean:", mean_a)
print("Variance:", variance_a)
print("Standard Deviation:", std_dev_a)

print("\nDataset B:")
print("Mean:", mean_b)
print("Variance:", variance_b)
print("Standard Deviation:", std_dev_b)
```
